```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split, validation_curve, StratifiedKFold
from sklearn.metrics import roc_auc_score
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10,6)
```

```python
train = pd.read_csv('03_santander_train.csv', index_col='ID')
```

```python
train.head()
```

```python
test = pd.read_csv('03_santander_test.csv', index_col='ID')
```

```python
test.head()
```

```python
y = list(train['TARGET'])
```

```python
X = train.drop('TARGET', axis=1)
```

```python
X.info()
```

```python
# выбираем 50 самых важных признаков для ускорения
model = DecisionTreeClassifier()
model.fit(X, y)
imp = pd.Series(index=X.columns, data=model.feature_importances_).sort_values(ascending=False).head(50).index.values
imp
```

```python
cv = StratifiedKFold(n_splits=3, random_state=123)
```

```python
model = DecisionTreeClassifier()
params = range(1, 11)
_, scores = validation_curve(model, X[imp], y, param_name='max_depth', param_range=params, 
                 scoring='roc_auc', cv=cv, n_jobs=-1)
plt.plot(params, scores.mean(axis=1))
```

```python
model = KNeighborsClassifier()
params = range(1, 21)
_, scores = validation_curve(model, X[imp], y, param_name='n_neighbors', param_range=params, 
                 scoring='roc_auc', cv=cv, n_jobs=-1)
plt.plot(params, scores.mean(axis=1))
```

```python
penalties = ['l1', 'l2']
Cs = [0.05, 0.1, 0.5, 1.0, 5]
for pen in penalties:
    model = LogisticRegression(penalty=pen)
    _, scores = validation_curve(model, X[imp], y, param_name='C', param_range=Cs, 
                         scoring='roc_auc', cv=cv, n_jobs=-1)
    plt.plot(np.log(Cs), scores.mean(axis=1), label=pen)
plt.legend()
```
